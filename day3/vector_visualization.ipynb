Display the point shapefile using the plot function

```python
pt_red.plot()
```

We can change the color size and shape of marker by using the appropriate argument in the plot function

```python
pt_red.plot(color='green')
```

```python
pt_red.plot(color='green',marker='D') # D is for diamond
```

```python
pt_red.plot(color='green',marker='v', markersize=5)
```

We can specify the colum if we would like to apply a colour map based on any column.

```python
pt_red.plot(column='STATE',cmap='viridis',marker='v', markersize=5,categorical=True,legend=True)
```

We can also change the marker size based on column name

```python
pt_red.plot(column='STATE',cmap='viridis',marker='v', markersize='howpopular',categorical=True)
```

We can also reorganize the markersizze based on the values in a column. Here the column in consideration has valuess from 0 to 5. But the markersize of 5 appears very small. So we can modify the column values by multiplying by 5 and populating this into a new column called popularity.

```python
pt_red['popularity']=5*pt_red['howpopular']
```

```python
pt_red.plot(column='STATE',cmap='viridis',marker='v', markersize='popularity',categorical=True)
```
